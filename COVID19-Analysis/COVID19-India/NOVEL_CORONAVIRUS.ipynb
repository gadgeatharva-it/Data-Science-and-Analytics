## Merging Data Frames

```python
df_final_India = pd.DataFrame()
dates = pd.DataFrame({"Date": pd.date_range(df_India.Date.min(),df_India.Date.max())})
for state in States:
    all_dates_df = pd.merge(dates, 
                            df_India.loc[df_India['State/UnionTerritory'] == state,:], on = "Date",
                           how = "left")
    all_dates_df['State/UnionTerritory'] = state
    all_dates_df = all_dates_df.fillna(0)
    all_dates_df['New Cases'] = all_dates_df['Confirmed'] - all_dates_df['Confirmed'].shift(1)
#     print(state)
#     display(all_dates_df.loc[all_dates_df['New Cases'] <  0,:])
    df_final_India = pd.concat([df_final_India, all_dates_df],axis = 0)
print("Finally we have a data of Size: ",df_final_India.shape)
df_final_India.head()
```

# Statewise Covid19 Status in India

```python
def plot_pie(active,cured,death,title):
    labels = ['Active','Recovered','Died']
    sizes = [active,cured,death]
    color= ['#66b3ff','green','red']
    explode = []

    for i in labels:
        explode.append(0.05)
    
    plt.figure(figsize= (15,6))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=9, explode =explode,colors = color)
    centre_circle = plt.Circle((0,0),0.70,fc='white')

    fig = plt.gcf()
    fig.gca().add_artist(centre_circle)
    plt.title(title +  'COVID-19 Cases',fontsize = 20)
    plt.axis('equal')  
    plt.tight_layout()
```

```python
total_cases_india = 0
cured_cases_india = 0
death_cases_india = 0
active_cases_india = 0
state_df = pd.DataFrame()

for state in States:
    one_state_df = df_final_India.loc[df_final_India['State/UnionTerritory'] == state,:]
    state_df = pd.concat([state_df,pd.DataFrame(one_state_df.iloc[-1,:]).T],axis = 0)
    total_cases = one_state_df['Confirmed'].values[-1]
    cured = one_state_df['Cured'].values[-1]
    deaths = one_state_df['Deaths'].values[-1]
    active = total_cases - cured - deaths
    plot_pie(active, cured, deaths,state)
    total_cases_india += total_cases
    cured_cases_india += cured
    death_cases_india += deaths
    active_cases_india += active
```

# Forecasting Using fbprophet

```python
# Run this cell first
df = df_daywise_India.iloc[:-1,]
df_train = df.loc[df['Date']<= "2020-05-23",:]
df_test = df.loc[df['Date'] > "2020-05-23",:]

confirmed_train = df_train[['Date','Confirmed']]
confirmed_test = df_test[['Date','Confirmed']]

deaths_train = df_train[['Date','Deaths']]
deaths_test = df_test[['Date','Deaths']]

recovered_train = df_train[['Date','Cured']]
recovered_test = df_test[['Date','Cured']]

confirmed_train.columns = ['ds','y']

# Now run the forecasting code
print("✅ Data ready!")
print(confirmed_train.tail())
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# ── Confirmed Cases Forecast ──
confirmed_train_copy = confirmed_train.copy()
confirmed_train_copy['day_num'] = np.arange(len(confirmed_train_copy))

X_train = confirmed_train_copy[['day_num']]
y_train = confirmed_train_copy['y']

model = LinearRegression()
model.fit(X_train, y_train)

# Forecast next 5 days
future_days = np.arange(len(confirmed_train_copy), len(confirmed_train_copy) + 5).reshape(-1, 1)
predictions = model.predict(future_days)

# Plot
plt.figure(figsize=(12, 5))
plt.plot(confirmed_train_copy['day_num'], y_train, label='Actual Confirmed', color='blue')
plt.plot(range(len(confirmed_train_copy), len(confirmed_train_copy) + 5), 
         predictions, label='Forecast', color='red', linestyle='--', marker='o')
plt.title('COVID-19 Confirmed Cases Forecast - India')
plt.xlabel('Days')
plt.ylabel('Confirmed Cases')
plt.legend()
plt.grid(True)
plt.show()

print("Next 5 days forecast:")
for i, val in enumerate(predictions):
    print(f"Day {i+1}: {int(val):,} cases")
```

```python
# Reset and rename all train/test dataframes
confirmed_train = df_train[['Date','Confirmed']].copy()
recovered_train = df_train[['Date','Cured']].copy()
deaths_train    = df_train[['Date','Deaths']].copy()

confirmed_test = df_test[['Date','Confirmed']].copy()
recovered_test = df_test[['Date','Cured']].copy()
deaths_test    = df_test[['Date','Deaths']].copy()

# Rename all to ds, y
confirmed_train.columns = ['ds','y']
recovered_train.columns = ['ds','y']
deaths_train.columns    = ['ds','y']

print("✅ Done!")
print(recovered_train.tail(3))
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

def forecast_linear(train_df, test_df, title):
    train = train_df.copy()
    train['day_num'] = np.arange(len(train))
    
    model = LinearRegression()
    model.fit(train[['day_num']], train['y'])
    
    future_days = np.arange(len(train), len(train) + 5).reshape(-1, 1)
    preds = model.predict(future_days)
    
    result = test_df.copy().reset_index(drop=True)
    result.columns = ['ds', 'Actual']
    result['yhat'] = preds[:len(result)]
    
    plt.figure(figsize=(10, 4))
    plt.plot(result['ds'], result['Actual'], color='red', marker='o', label='Actual', alpha=0.6)
    plt.plot(result['ds'], result['yhat'], color='green', marker='o', label='Predicted', alpha=0.6, linestyle='--')
    plt.title(title)
    plt.xlabel('Date')
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()

# ── Run all 3 ──
forecast_linear(confirmed_train, confirmed_test, 'Results - Total Cases')
forecast_linear(recovered_train, recovered_test, 'Results - Recovered')
forecast_linear(deaths_train,    deaths_test,    'Results - Deaths')
```
