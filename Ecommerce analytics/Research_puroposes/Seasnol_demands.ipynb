```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_excel(r"C:\Users\sudhi\Desktop\Reasearch_dataset\ecommerce_data.xlsx")

# Clean columns
data.columns = data.columns.str.strip().str.lower()

# Convert date
data['transaction_date'] = pd.to_datetime(data['transaction_date'])

# Convert price
data['product_price'] = pd.to_numeric(data['product_price'], errors='coerce')

# 🔥 Extract month
data['month'] = data['transaction_date'].dt.month

# Group by month (average demand)
monthly_demand = data.groupby('month')['product_price'].mean()

# Month labels
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

# Plot
plt.figure()

plt.plot(monthly_demand.index, monthly_demand.values, marker='o')

plt.xticks(range(1,13), months)

plt.xlabel("Month")
plt.ylabel("Average Demand")
plt.title("Seasonal Demand Pattern in E-Commerce")

plt.tight_layout()
plt.show()
```
