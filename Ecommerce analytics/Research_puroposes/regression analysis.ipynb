```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load data
data = pd.read_excel(r"C:\Users\sudhi\Desktop\Reasearch_dataset\ecommerce_data.xlsx")

# Clean columns
data.columns = data.columns.str.strip().str.lower()

# Fix column name (important)
data.rename(columns={'standerd_price': 'standard_price'}, inplace=True)

# Convert numeric
data['product_price'] = pd.to_numeric(data['product_price'], errors='coerce')
data['standard_price'] = pd.to_numeric(data['standard_price'], errors='coerce')

# Drop missing values
data = data.dropna(subset=['product_price', 'standard_price'])

# =========================
# Regression Model
# =========================
X = data[['standard_price']]   # independent variable
y = data['product_price']      # dependent variable

model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# =========================
# Plot
# =========================
plt.figure()

plt.scatter(X, y, label="Actual Data")
plt.plot(X, y_pred, linestyle='dashed', label="Regression Line")

plt.xlabel("Standard Cost")
plt.ylabel("Product Price")
plt.title("Regression Analysis: Price vs Cost")

plt.legend()
plt.tight_layout()
plt.show()
```
