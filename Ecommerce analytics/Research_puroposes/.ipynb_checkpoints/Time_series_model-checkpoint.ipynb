```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your file
data = pd.read_excel(r"C:\Users\sudhi\Desktop\Reasearch_dataset\ecommerce_data.xlsx")

# Clean column names
data.columns = data.columns.str.strip()

# Convert price
data['product_price'] = pd.to_numeric(data['product_price'], errors='coerce')

# 🔥 Segment by category
category_segment = data.groupby('category')['product_price'].mean()

# Sort for better visualization
category_segment = category_segment.sort_values(ascending=False)

print(category_segment)
```
