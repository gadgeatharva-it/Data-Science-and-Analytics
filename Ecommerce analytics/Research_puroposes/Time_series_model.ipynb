```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_excel(r"C:\Users\sudhi\Desktop\Reasearch_dataset\ecommerce_data.xlsx")

# Clean column names
data.columns = data.columns.str.strip().str.lower()

# Convert date
data['transaction_date'] = pd.to_datetime(data['transaction_date'])

# Convert price
data['product_price'] = pd.to_numeric(data['product_price'], errors='coerce')

# 🔥 Set date as index
data.set_index('transaction_date', inplace=True)

# 🔥 Monthly demand (average)
monthly_demand = data['product_price'].resample('M').mean()

# Plot
plt.figure()

plt.plot(monthly_demand.index, monthly_demand.values, marker='o')

plt.title("Monthly Demand Trend in E-Commerce")
plt.xlabel("Date")
plt.ylabel("Average Demand")

plt.xticks(rotation=45)
plt.grid()

plt.tight_layout()
plt.show()
```
