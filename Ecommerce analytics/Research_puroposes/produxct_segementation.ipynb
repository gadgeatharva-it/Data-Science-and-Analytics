```python
import pandas as pd
import matplotlib.pyplot as plt

# =========================
# 1. LOAD DATA
# =========================
data = pd.read_excel(r"C:\Users\sudhi\Desktop\Reasearch_dataset\ecommerce_data.xlsx")   # keep file in same folder

# =========================
# 2. CLEAN DATA
# =========================
data.columns = data.columns.str.strip().str.lower()

# Convert price to numeric
data['product_price'] = pd.to_numeric(data['product_price'], errors='coerce')

# =========================
# 3. DEMAND CALCULATION
# =========================
# Use SUM for demand (better than mean)
category_demand = data.groupby('category')['product_price'].sum()

# Sort values
category_demand = category_demand.sort_values(ascending=False)

# =========================
# 4. DISPLAY RESULT
# =========================
print("Predicted Demand by Category:\n")
print(category_demand)

# =========================
# 5. PLOT GRAPH
# =========================
plt.figure()

plt.bar(category_demand.index, category_demand.values)

plt.xlabel("Product Category")
plt.ylabel("Total Sales (Demand)")
plt.title("Predicted Demand by Product Category")

plt.xticks(rotation=30)
plt.tight_layout()

plt.show()
```
